```python
# Run Inference UI
from ipywidgets import Layout

# style to fix too long descriptions
style = {'description_width': 'initial'}

# Define the widgets for each parameter
prompt_widget = widgets.Text(
    value='An agent',
    description='Prompt:',
    style=style
)

stable_diffusion_model_widget = widgets.Dropdown(
    options=['2', 'xl'],
    value='xl',
    description='Stable Diffusion Model (xl understands better prompts):',
    style=style,
    layout=Layout(width='450px')
)

num_inference_steps_widget = widgets.IntText(
    value=25,
    min=1,
    description='Number of Inference Steps (more value = better image but longer process):',
    orientation='horizontal',
    readout=True,
    readout_format='d',
    style=style,
    layout=Layout(width='550px')
)

guidance_scale_widget = widgets.FloatText(
    value=7.5,
    min=0.1,
    description='Guidance Scale (more value = more adherent to the prompt):',
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    style=style,
    layout=Layout(width='600px')
)

model_precision_type_widget = widgets.Dropdown(
    options=['fp16', 'fp32'],
    value='fp16',
    description='Model Precision Type (fp16 is faster but fp32 is more precise but slower and requires paid colab gpu):',
    style=style,
    layout=Layout(width='700px')
)

seed_widget = widgets.IntText(
    value=0,
    description='Seed (A starting point to initate generation, put 0 for random):',
    readout=True,
    readout_format='d',
    style=style,
    layout=Layout(width='600px')
)

filename_widget = widgets.Text(
    value='output-skin.png',
    description='Output Image Name:',
    style=style,
    layout=Layout(width='600px')
)

model_3d_widget = widgets.Checkbox(
    value=True,
    description="See the skin also as a 3D Model",
    indent=False,
    style=style,
    layout=Layout(width='600px')
)

verbose_widget = widgets.Checkbox(
    value=False,
    description="Verbose (more detailed output info)",
    indent=False,
    style=style,
    layout=Layout(width='600px')
)

# Create a function to handle the button click event
def run_inference(b):
    prompt = prompt_widget.value
    stable_diffusion_model = stable_diffusion_model_widget.value
    num_inference_steps = num_inference_steps_widget.value
    guidance_scale = guidance_scale_widget.value
    model_precision_type = model_precision_type_widget.value
    seed = seed_widget.value
    filename = filename_widget.value
    verbose = verbose_widget.value
    model_3d = model_3d_widget.value

    if verbose:
        verbose_opt = '--verbose'
    else:
        verbose_opt = ''

    if model_3d:
        model_3d_opt = '--model_3d'
    else:
        model_3d_opt = ''

    if stable_diffusion_model == '2':
        sd_model = "minecraft-skins"
    else:
        sd_model = "minecraft-skins-sdxl"

    command = f"Scripts/{sd_model}.py '{prompt}' {num_inference_steps} {guidance_scale} {model_precision_type} {seed} {filename} {model_3d_opt} {verbose_opt}"

    !python $command


def display_asset(b):

  filename = filename_widget.value

  import matplotlib.pyplot as plt

  img = plt.imread(f"output_minecraft_skins/{filename}")

  plt.axis("off")

  plt.imshow(img)


# Create a button widget and attach the click event handler
run = widgets.Button(
    description="Run Inference (Generate Skin)",
    layout=Layout(width='600px'))

run.on_click(run_inference)


# display asset output
display_asset_button = widgets.Button(
    description="Display Skin Asset Output",
    layout=Layout(width='600px'))

display_asset_button.on_click(display_asset)


# Display the widgets and the button
display(prompt_widget)
display(stable_diffusion_model_widget)
display(num_inference_steps_widget)
display(guidance_scale_widget)
display(model_precision_type_widget)
display(seed_widget)
display(filename_widget)
display(model_3d_widget)
display(verbose_widget)
display(run)
display(display_asset_button)
```

# HOW TO DOWNLOAD OUTPUT SKIN (ASSET & 3D MODEL)
After you click Run Inference on the Ipywidgets UI, open the Console at the bottom and you should see the progress of the skin, when you see 'Successfully generated the skin!' it means it's done!
At your right you should find 'Output', expand the /kaggle/working folder and then the Minecraft_Skin_Generator folder, you will find the output asset image in the 'output_minecraft_skins' folder with the name you gave it, and also you can find the .glb 3d model if you made model_3d = True in the inference cell before, there will be a (the name you gave it)-converted.png too that was just made to fix a bug to load the texture on the 3d model in the project files (this bug won't happen in-game), just click the 3 dots at its right of it and click download
