```python
#@title Run Inference UI
from ipywidgets import Layout

# style to fix too long descriptions
style = {'description_width': 'initial'}

# Define the widgets for each parameter
prompt_widget = widgets.Text(
    value='An agent',
    description='Prompt:',
    style=style
)

stable_diffusion_model_widget = widgets.Dropdown(
    options=['2', 'xl'],
    value='xl',
    description='Stable Diffusion Model (xl understands better prompts):',
    style=style,
    layout=Layout(width='450px')
)

num_inference_steps_widget = widgets.IntText(
    value=50,
    min=1,
    description='Number of Inference Steps (more value = better image but longer process):',
    orientation='horizontal',
    readout=True,
    readout_format='d',
    style=style,
    layout=Layout(width='550px')
)

guidance_scale_widget = widgets.FloatText(
    value=7.5,
    min=0.1,
    description='Guidance Scale (more value = more adherent to the prompt):',
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    style=style,
    layout=Layout(width='600px')
)

model_precision_type_widget = widgets.Dropdown(
    options=['fp16', 'fp32'],
    value='fp16',
    description='Model Precision Type (fp16 is faster but fp32 is more precise but slower and requires paid colab gpu):',
    style=style,
    layout=Layout(width='700px')
)

seed_widget = widgets.IntText(
    value=0,
    description='Seed (A starting point to initate generation, put 0 for random):',
    readout=True,
    readout_format='d',
    style=style,
    layout=Layout(width='600px')
)

output_image_name_widget = widgets.Text(
    value='output-skin.png',
    description='Output Image Name:',
    style=style,
    layout=Layout(width='600px')
)

verbose_widget = widgets.Checkbox(
    value=False,
    description="Verbose (more detailed output info)",
    indent=False,
    style=style,
    layout=Layout(width='600px')
)

# Create a function to handle the button click event
def run_inference(b):
    prompt = prompt_widget.value
    stable_diffusion_model = stable_diffusion_model_widget.value
    num_inference_steps = num_inference_steps_widget.value
    guidance_scale = guidance_scale_widget.value
    model_precision_type = model_precision_type_widget.value
    seed = seed_widget.value
    output_image_name = output_image_name_widget.value
    verbose = verbose_widget.value

    if verbose:
        verbose_opt = '--verbose'
    else:
        verbose_opt = ''

    if stable_diffusion_model == '2':
        sd_model = "minecraft-skins"
    else:
        sd_model = "minecraft-skins-sdxl"

    command = f"Python_Scripts/{sd_model}.py '{prompt}' {num_inference_steps} {guidance_scale} {model_precision_type} {seed} {output_image_name} {verbose_opt}"

    !python $command

    files.download(f"output_minecraft_skins/{output_image_name}")


# Create a button widget and attach the click event handler
button = widgets.Button(
    description="Run Inference (Generate Skin)",
    layout=Layout(width='600px'))

button.on_click(run_inference)

# Display the widgets and the button
display(prompt_widget)
display(stable_diffusion_model_widget)
display(num_inference_steps_widget)
display(guidance_scale_widget)
display(model_precision_type_widget)
display(seed_widget)
display(output_image_name_widget)
display(verbose_widget)
display(button)
```
